```python
import pandas as pd
import datetime as datetime
import numpy as np
from dateutil.relativedelta import relativedelta
```

```python
adm_icu = pd.DataFrame.from_csv('ADMISSIONS_ICUSTAY_SEPSIS.csv')
adm_icu.head()
```

```python
dates_and_times = ['dob', 'dod', 'admittime', 'dischtime', 'deathtime', 'intime', 'outtime']
for thing in dates_and_times:
    adm_icu[thing] = pd.to_datetime(adm_icu[thing])
    
adm_icu.head()
```

```python
''' this code was written to re-calculate the ages of patients after the 
    data was already calculated by different means. This code addresses NaT 
    values generated below that were introduced to replace age < 0
    the code uses relativedelta.years to get the age difference rounded years'''
for index, row in adm_icu.iterrows():
    if type(row['dod']) != pd.tslib.NaTType:
        row['age'] = relativedelta(row['dod'], row['dob']).years

#this is a way to check to see if dates are NaT/nan
#type(row['dod']) == pd.tslib.NaTType

#calculate the duration of hospital stay and duration of icu stays. 
adm_icu['admission_stay'] = adm_icu['dischtime'] - adm_icu['admittime']
adm_icu['icu_stay'] = adm_icu['outtime'] - adm_icu['intime']
adm_icu.head()
```
